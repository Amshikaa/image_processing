```python
import os
import cv2
import matplotlib.pyplot as plt

def find_mean_variance(img):
    rows, cols = img.shape
    
    sum1 = 0

    for i in range(rows):
        for j in range(cols):
            sum1 += img[i,j]
    mean = sum1 / (rows * cols)
    
    varianceSum = 0
    
    for i in range(rows):
        for j in range(cols):
            varianceSum += (img[i,j] - mean) ** 2
    variance = varianceSum / (rows * cols)
    
    return mean, variance 
    
root_folder = "C:\\Users\\User\\image_processing\\image_processing\\20_images_folder"
imgList = []
for file in os.listdir(root_folder): 
    if file.endswith(('.jpg', '.jpeg', '.png', '.jfif')):
        file_path = os.path.join(root_folder,file)
        imgList.append(file_path)

means = []
variances = []

for i in range(len(imgList)):
    print(imgList[i])
    imgOb = cv2.imread(imgList[i]) 
    
    gray_image = cv2.cvtColor(imgOb, cv2.COLOR_BGR2GRAY) 
    mean, variance = find_mean_variance(gray_image)
    
    means.append(mean)
    variances.append(variance)

print("Means: ", means)
print("Variances : ", variances)
plt.scatter(means,variances)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load images
image1 = cv2.imread('im1.jpg')
image2 = cv2.imread('im2.jpg')
rows, cols = image2.shape[:2]

# Ensure the images are of the same size
image1 = cv2.resize(image1, (cols,rows))

added_image = cv2.add(image1, image2)
sub_image = cv2.subtract(image2, image1)
multiplied_image = cv2.multiply(image1, image2)

image1_float = image1.astype(np.float32)
image2_float = image2.astype(np.float32) + 1  # Adding 1 to avoid division by zero

divided_image = cv2.divide(image1_float, image2_float)

# Convert back to uint8
divided_image = cv2.normalize(divided_image, None, 0, 255, cv2.NORM_MINMAX)
divided_image = divided_image.astype(np.uint8)

plt.figure(figsize = (8,12))
plt.subplot(3,2,1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(3,2,2)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(3,2,3)
plt.imshow(cv2.cvtColor(added_image, cv2.COLOR_BGR2RGB))
plt.title('Added Image')

plt.subplot(3,2,4)
plt.imshow(cv2.cvtColor(sub_image, cv2.COLOR_BGR2RGB))
plt.title('Subtracted Image')

plt.subplot(3,2,5)
plt.imshow(cv2.cvtColor(multiplied_image, cv2.COLOR_BGR2RGB))
plt.title('Multiplied Image')

plt.subplot(3,2,6)
plt.imshow(cv2.cvtColor(divided_image, cv2.COLOR_BGR2RGB))
plt.title('Divided Image')
```
