```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the five images
img1 = cv2.imread('im1.jpg')
img2 = cv2.imread('im2.jpg')
img3 = cv2.imread('im3.jpg')
img4 = cv2.imread('im4.jpg')
img5 = cv2.imread('im5.jpg')

# Get the dimensions of the first image
height, width, channels = img1.shape

# Resize the images to match the dimensions of the first image
img2 = cv2.resize(img2, (width, height))
img3 = cv2.resize(img3, (width, height))
img4 = cv2.resize(img4, (width, height))
img5 = cv2.resize(img5, (width, height))

# Convert images to numpy arrays
images = [img1, img2, img3, img4, img5]

# Create subplots
fig, axs = plt.subplots(1, 6, figsize=(24, 4))

# Loop through each image and display it in a subplot
for i, image in enumerate(images):
    # Convert from BGR to RGB if necessary
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axs[i].imshow(image_rgb)
    axs[i].axis('off')

# Calculate the average image
average_image = np.mean(images, axis=0).astype(np.uint8)
average_image_rgb = cv2.cvtColor(average_image, cv2.COLOR_BGR2RGB)

# Display the average image
axs[-1].imshow(average_image_rgb)
axs[-1].set_title('Average Image')
axs[-1].axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread('lab1.jpg')
img2=cv2.imread('lab2.jpg')

rows, cols = img2.shape[:2]

img1=cv2.resize(img1,(cols,rows))
diff=cv2.subtract(img2,img1)

plt.subplot(1, 3, 1) 
plt.title("Original image1") 
plt.imshow(img1)

plt.subplot(1, 3, 2) 
plt.title("Original image2") 
plt.imshow(img2)

plt.subplot(1, 3, 3) 
plt.title("subtracted image") 
plt.imshow(diff)
plt.show()
```
