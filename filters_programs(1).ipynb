```python
import numpy as np
import cv2

# Define the kernel for filtering
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Read the image
image = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)

# Get the dimensions of the image and the kernel
image_height, image_width = image.shape
kernel_height, kernel_width = kernel.shape

# Define a function to perform spatial correlation
def spatial_correlation(image, kernel):
    result = np.zeros_like(image)
    for y in range(image_height):
        for x in range(image_width):
            sum = 0
            for ky in range(kernel_height):
                for kx in range(kernel_width):
                    img_y = y - kernel_height // 2 + ky
                    img_x = x - kernel_width // 2 + kx
                    if img_y >= 0 and img_y < image_height and img_x >= 0 and img_x < image_width:
                        sum += image[img_y, img_x] * kernel[ky, kx]
            result[y, x] = np.clip(sum, 0, 255)
    return result

# Define a function to perform convolution
def convolution(image, kernel):
    result = np.zeros_like(image)
    kernel = np.flipud(np.fliplr(kernel))
    for y in range(image_height):
        for x in range(image_width):
            sum = 0
            for ky in range(kernel_height):
                for kx in range(kernel_width):
                    img_y = y - kernel_height // 2 + ky
                    img_x = x - kernel_width // 2 + kx
                    if img_y >= 0 and img_y < image_height and img_x >= 0 and img_x < image_width:
                        sum += image[img_y, img_x] * kernel[ky, kx]
            result[y, x] = np.clip(sum, 0, 255)
    return result

# Perform spatial correlation
correlation_result = spatial_correlation(image, kernel)

# Perform convolution
convolution_result = convolution(image, kernel)

# Display the results
cv2.imshow('Original Image', image)
cv2.imshow('Correlation Result', correlation_result)
cv2.imshow('Convolution Result', convolution_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Average filter
import numpy as np
import cv2

# Read the image
image = cv2.imread('dog.jpg', cv2.IMREAD_GRAYSCALE)

# Get the dimensions of the image
image_height, image_width = image.shape

# Define the kernel size for average filtering
kernel_size = 3

# Define a function to perform average filtering
def average_filter(image, kernel_size):
    result = np.zeros_like(image)
    for y in range(image_height):
        for x in range(image_width):
            sum = 0
            count = 0
            for ky in range(-(kernel_size // 2), kernel_size // 2 + 1):
                for kx in range(-(kernel_size // 2), kernel_size // 2 + 1):
                    img_y = y + ky
                    img_x = x + kx
                    if img_y >= 0 and img_y < image_height and img_x >= 0 and img_x < image_width:
                        sum += image[img_y, img_x]
                        count += 1
            result[y, x] = sum // count
    return result

# Perform average filtering
filtered_image = average_filter(image, kernel_size)

# Display the results
cv2.imshow('Original Image', image)
cv2.imshow('Filtered Image', filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Max filter
import numpy as np
import cv2

# Read the image
image = cv2.imread('dog.jpg', cv2.IMREAD_GRAYSCALE)

# Get the dimensions of the image
image_height, image_width = image.shape

# Define the kernel size for max filtering
kernel_size = 3

# Define a function to perform max filtering
def max_filter(image, kernel_size):
    result = np.zeros_like(image)
    for y in range(image_height):
        for x in range(image_width):
            max_val = 0
            for ky in range(-(kernel_size // 2), kernel_size // 2 + 1):
                for kx in range(-(kernel_size // 2), kernel_size // 2 + 1):
                    img_y = y + ky
                    img_x = x + kx
                    if img_y >= 0 and img_y < image_height and img_x >= 0 and img_x < image_width:
                        max_val = max(max_val, image[img_y, img_x])
            result[y, x] = max_val
    return result

# Perform max filtering
filtered_image = max_filter(image, kernel_size)

# Display the results
cv2.imshow('Original Image', image)
cv2.imshow('Filtered Image', filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Min filter
import numpy as np
import cv2

# Read the image
image = cv2.imread('dog.jpg', cv2.IMREAD_GRAYSCALE)

# Get the dimensions of the image
image_height, image_width = image.shape

# Define the kernel size for min filtering
kernel_size = 3

# Define a function to perform min filtering
def min_filter(image, kernel_size):
    result = np.zeros_like(image)
    for y in range(image_height):
        for x in range(image_width):
            min_val = 255  # Assuming 8-bit grayscale image
            for ky in range(-(kernel_size // 2), kernel_size // 2 + 1):
                for kx in range(-(kernel_size // 2), kernel_size // 2 + 1):
                    img_y = y + ky
                    img_x = x + kx
                    if img_y >= 0 and img_y < image_height and img_x >= 0 and img_x < image_width:
                        min_val = min(min_val, image[img_y, img_x])
            result[y, x] = min_val
    return result

# Perform min filtering
filtered_image = min_filter(image, kernel_size)

# Display the results
cv2.imshow('Original Image', image)
cv2.imshow('Filtered Image', filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

image = cv2.imread('nature.jpg', cv2.IMREAD_GRAYSCALE)

image_height, image_width = image.shape

kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

def convolution(image, kernel):
    result = np.zeros_like(image)
    kernel = np.flipud(np.fliplr(kernel))
    for y in range(image_height):
        for x in range(image_width):
            sum = 0
            for ky in range(kernel.shape[0]):
                for kx in range(kernel.shape[1]):
                    img_y = y - kernel.shape[0] // 2 + ky
                    img_x = x - kernel.shape[1] // 2 + kx
                    if img_y >= 0 and img_y < image_height and img_x >= 0 and img_x < image_width:
                        sum += image[img_y, img_x] * kernel[ky, kx]
            result[y, x] = np.clip(sum, 0, 255)
    return result

sharpened_image = convolution(image, kernel)

cv2.imshow('Original Image', image)
cv2.imshow('Sharpened Image', sharpened_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
