```python
from PIL import Image
def compute_average(image_path):
    image = Image.open(image_path)
    width, height = image.size
    
    sum_pixels = 0
    num_pixels = 0
    
    for y in range(height):
        for x in range(width):
            
            r, g, b = image.getpixel((x, y))
            
            sum_pixels += (r + g + b) / 3
            num_pixels += 1
    
    average_pixel_value = sum_pixels / num_pixels
    
    return average_pixel_value

image_path = 'dog.jpg'
average_value = compute_average(image_path)
print(f'The average pixel value of the image is: {average_value}')
```

```python
#Python program to find the average of grayscale image without using built in functions
def average(image):
    sum = 0
    count = 0
    for row in image:
        for pixel in row:
            sum += pixel
            count += 1
    average = sum / count
    return average
image = [
    [10, 20, 10],
    [10, 20, 10],
    [7, 3, 5]
]
average = average(image)
print("Average intensity of the grayscale image:", average)
```

```python
#Python program to find the average and variance of an image without using built in functions
import math
import cv2
def mean_variance(image):
    sums = 0
    sum_squared = 0
    count = 0
    for row in image:
        for pixel in row:
            sums += pixel
            sum_squared += pixel ** 2
            count += 1
    mean = sums / count      
    variance = mean**(0.5)
    return mean
    return variance
image = cv2.imread("dog.jpg",0)
mean = mean_variance(image)
variance = mean_variance(image)
print("Average of the grayscale image:", mean)
print("Variance of the grayscale image:", variance)
```

```python
#Euclidean Distance
import math
v1=[4,6,8]
v2=[8,7,2]
sum=0
for i in range(len(v1)):
    sum+=(v1[i]-v2[i])**2
ED=math.sqrt(sum)
print("ED:",ED)
```

```python
import cv2
image=input("Enter the image")
img = cv2.imread(image, 0)
print("Original matrix\n",img)

rows, cols = img.shape

sum1 = 0

for i in range(rows):
    for j in range(cols):
        sum1 += img[i,j]
mean = sum1 / (rows * cols)
print("Mean :: ", mean)

varianceSum = 0
for i in range(rows):
    for j in range(cols):
        varianceSum += (img[i,j] - mean) ** 2
variance = varianceSum / (rows * cols)
print("Variance :: ", variance)
```
